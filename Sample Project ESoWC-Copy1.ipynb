```python
import numpy as np 
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure,output_file,show
from bokeh.models import ColumnDataSource,ColorBar,LogColorMapper
from math import pi
```

```python
output_notebook()  
```

```python
df = pd.read_csv('/home/bhavya/Desktop/example.csv')

source = ColumnDataSource(df)

output_file("Average_times_of_Routines.html")
```

```python
#Figure layout for Render 1:

TOOLTIPS = [('routine', "@routine"),('Time', '@avgTime')]

TOOLS = "tap,pan,wheel_zoom,reset,hover,save"

p = figure(x_range=df["routine"],title = "Average times of Routines",sizing_mode = "stretch_both",tools = TOOLS,tooltips=TOOLTIPS)
p.background_fill_color = "beige"
p.border_fill_color = "whitesmoke"
p.outline_line_width = 7
p.outline_line_alpha = 0.3
p.outline_line_color = "brown"
p.xaxis.axis_label = "Routines"
p.yaxis.axis_label = "Average Times"
p.xaxis.major_label_orientation = pi/4
```

```python
#Render 1 : Bar graph

color_mapper = LogColorMapper(palette="OrRd9", low=5, high=36)

r = p.vbar(x = 'routine', top = 'avgTime',width=0.9,source = source,fill_color={'field': 'avgTime', 'transform': color_mapper})
glyph = r.glyph
glyph.line_color = "firebrick"
glyph.fill_alpha = 0.8
color_bar = ColorBar(color_mapper=color_mapper, 
                     label_standoff=12, border_line_color=None, location=(0,0))

p.add_layout(color_bar, 'right')
p.toolbar.autohide = True

show(p)
```
